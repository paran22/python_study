## 📄 프로그래밍 파일의 정체: `.py` vs `.ipynb`

| | `.py` (스크립트 파일) | `.ipynb` (노트북 파일) |
|---|---|---|
| **실행 방식** | 위에서 아래로 전체를 한 번에 실행 | 셀(cell) 단위로 한 블록씩 실행 |

| **출력** | `print()`를 써야 출력됨 | 셀의 마지막 줄 결과가 자동으로 출력됨 (`print` 없이도 확인 가능) |
| **작업 환경** | 로컬 IDE에서 작성하고 실행 | Google Colab 등 노트북 환경 |

## Part 2. 따옴표와 괄호의 비밀

- **따옴표(`" "` 또는 `' '`)** 안에 있으면 → 글자(텍스트 데이터)
- **따옴표 없이** 쓰면 → 파이썬이 명령어나 변수 이름으로 인식
- **괄호**는 반드시 열면 닫아야 합니다


````python
# 따옴표 있음 → 글자 데이터
print("hello")   # hello를 출력
print('hello')   # 작은따옴표도 동일
````

````python
# 따옴표가 없으면? → 에러!
print(hello)     # hello라는 변수를 찾으려 함 → NameError
````

````python
# 따옴표 없음 → 변수로 인식
message = "안녕하세요"
print(message)   # 변수 message에 담긴 값을 출력
````

````python
# 괄호를 열었으면 반드시 닫아야 합니다
# 아래 주석을 해제하고 실행해보세요 — 에러가 납니다
# print("괄호를 안 닫으면"
````

## Part 3. 띄어쓰기 vs 들여쓰기

- **띄어쓰기**: 변수 이름, 연산자 사이 등은 자유롭게 써도 됩니다
- **들여쓰기(Tab 또는 스페이스 4칸)**: `if`, `for`, `def` 등 블록 안에서는 **필수**입니다. 틀리면 에러!


````python
# 들여쓰기가 올바른 경우
score = 90

if score >= 80:
    print("합격")   # if 블록 안은 들여쓰기 필수
````

````python
# 들여쓰기가 없으면 에러가 납니다
# 아래 주석을 해제하고 실행해보세요
# if score >= 80:
# print("합격")   # 들여쓰기 없음 → IndentationError!
````

### 7-3. 반환값(Return)

위 함수들은 `print`로 화면에 출력만 하고 끝났습니다.  
그런데 만약 함수가 만든 결과를 **저장하거나 다른 곳에서 다시 쓰고 싶다면** 어떻게 할까요?

이때 `return`을 사용합니다.

- `print` → 화면에 **보여주기만** 한다 (결과를 돌려주지 않음)
- `return` → 결과를 **돌려준다** (변수에 담아서 재사용 가능)

````python
# print만 있는 함수 — 화면에 보여주고 끝
def greet_print(name):
    print(f"{name}님, 반갑습니다!")

result = greet_print("가은")
print(result)  # None — 돌려받은 값이 없다!
````

````python
# return이 있는 함수 — 결과를 돌려준다
def greet_return(name):
    return f"{name}님, 반갑습니다!"

result = greet_return("가은")
print(result)  # 돌려받은 값이 있다!
````

````python
# return이 있으면 결과를 저장해서 다른 곳에서 재사용할 수 있습니다
greeting1 = greet_return("가은")
greeting2 = greet_return("민수")

# 두 결과를 합쳐서 새로운 문자열 만들기
combined = greeting1 + " " + greeting2
print(combined)
````

````python
# ✏️ 직접 해보기
# 빈칸을 채워서 return이 있는 함수를 완성해보세요

def make_message(name, topic):
    message = f"{name}님, {topic} 공부 화이팅!"
    return ""       # 무엇을 돌려줘야 할까요? ""를 지우고 알맞은 변수를 넣어보세요

result = make_message("가은", "파이썬")
print(result)
````

### 7-4. 변수가 살아있는 범위 (Scope)

함수 안에서 만든 변수는 함수 밖에서 사용할 수 없습니다.  
이를 **지역 변수**라고 합니다.

````python
# 함수 밖에서 만든 변수 → 함수 안에서도, 밖에서도 쓸 수 있음
language = "한국어"

def show_language():
    print(language)

show_language()  # 함수 안에서 출력
print(language)  # 함수 밖에서 출력
````

````python
# 함수 안에서 만든 변수 → 함수 밖에서 쓸 수 없음
def make_secret():
    secret = "함수 안에서만 존재하는 변수"
    print(secret)

make_secret()
````

````python
print(secret)  # NameError — 함수 밖에서는 접근 불가
````

## ✏️ 종합 실습: 프롬프트 생성 함수 만들기

오늘 배운 변수, f-string, 함수를 모두 활용해서 나만의 프롬프트 생성 함수를 만들어봅시다.

````python
# 예시: 프롬프트 생성 함수
def make_prompt(role, topic, language, style):
    # """ 안의 텍스트는 들여쓰기하면 출력에 불필요한 공백이 포함됩니다
    # 그래서 의도적으로 줄 맨 앞에 붙여 씁니다
    prompt = f"""
당신은 {role}입니다.
{topic}에 대해 {language}로 {style} 설명해주세요.
"""
    return prompt

# 함수를 호출해서 프롬프트 생성
result = make_prompt("IT 전문가", "인공지능", "한국어", "쉽고 친절하게")
print(result)
````

````python
# 입력값만 바꾸면 다양한 프롬프트를 만들 수 있습니다
print(make_prompt("영양사", "다이어트 식단", "한국어", "간결하게"))
print(make_prompt("여행 가이드", "도쿄 여행", "영어", "재미있게"))
````

````python
# ✏️ 직접 해보기
# 빈칸("")을 채워서 나만의 프롬프트 생성 함수를 완성해보세요

def my_prompt_maker(name, topic):
    role = ""           # 어떤 역할? (예: "마케팅 전문가")
    language = ""       # 어떤 언어? (예: "한국어")
    prompt = f"""
{role}의 입장에서 답변해주세요.
{name}님이 궁금해하는 주제: {topic}
답변 언어: {language}
"""
    return prompt

# 함수를 호출해서 결과 확인
result = my_prompt_maker("가은", "AI 에이전트")
print(result)
````

## Part 8. 독립된 작업실 만들기 (가상환경 세팅)

### 가상환경이란?

파이썬 프로젝트를 할 때는 다른 사람이 만든 패키지(도구 모음)를 설치해서 사용합니다.  
그런데 프로젝트마다 필요한 패키지와 **버전**이 다릅니다.

```
프로젝트 A: requests 2.28 필요
프로젝트 B: requests 2.31 필요
→ 같은 컴퓨터에서 충돌!
```

**가상환경(venv)**은 프로젝트마다 독립된 공간을 만들어서 이런 충돌을 방지합니다.  
프로젝트별로 따로 패키지를 설치하는 "전용 작업실"이라고 생각하면 됩니다.

### 실습은 Cursor에서!

아래 명령어들은 **Cursor 터미널**에서 직접 실행합니다.  
이 노트북은 참고용으로만 읽어보세요.

### 8-1. 가상환경 만들기와 활성화

Cursor 터미널에서 아래 명령어를 순서대로 실행합니다.

```bash
# 1. 가상환경 생성
# python -m venv <이름> → ".venv"라는 이름의 가상환경 폴더를 만듦
# -m venv: 파이썬에 내장된 가상환경 모듈을 실행하라는 뜻
python -m venv .venv

# 2. 가상환경 활성화
# 활성화해야 이 가상환경 안의 파이썬과 패키지를 사용하게 됨
source .venv/bin/activate        # Mac
.venv\Scripts\activate           # Windows (cmd에서 실행)

# 3. 터미널 앞에 (.venv)가 표시되면 성공!
# (.venv) ~/my-project $

# 4. 가상환경 비활성화 (작업 끝났을 때)
# 다시 원래 시스템 파이썬으로 돌아감
deactivate
```

### 8-2. 패키지 설치 (`pip`)

`pip`은 파이썬 패키지를 설치하는 명령어입니다.  
가상환경을 활성화한 상태에서 실행하면, 이 프로젝트 전용으로 설치됩니다.

```bash
# 패키지 설치
# pip install <패키지이름> → 최신 버전을 자동으로 설치
pip install requests

# 특정 버전을 지정해서 설치
# ==으로 정확한 버전을 고정할 수 있음
pip install requests==2.31.0

# 설치된 패키지 목록 확인
pip list

# 특정 패키지 삭제
pip uninstall requests
```

**버전이 중요한 이유:**  
같은 패키지라도 버전에 따라 동작이 달라질 수 있습니다.  
팀원 간 버전이 다르면 "내 컴퓨터에선 되는데?" 문제가 발생합니다.

### `requests` 패키지란?

파이썬에서 **웹 요청(HTTP)**을 보내는 패키지입니다.  
API를 호출하거나 웹에서 데이터를 가져올 때 사용합니다.  
3차시에서 Tavily 검색 API, LLM API를 호출할 때 이 패키지를 사용하게 됩니다.

```python
import requests

# 외부 API에 요청을 보내고 응답을 받아옴
response = requests.get("https://catfact.ninja/fact")
data = response.json()

print("응답 데이터:", data)
print("고양이 사실:", data["fact"])
```

### 8-3. `requirements.txt` — 패키지 목록 저장

설치된 패키지와 버전 번호를 파일로 저장해두면,  
다른 사람(또는 미래의 나)이 동일한 환경을 한 줄로 복원할 수 있습니다.

```bash
# 현재 설치된 패키지 목록을 파일로 저장
# > 는 출력 결과를 파일에 쓰라는 뜻
pip freeze > requirements.txt

# 저장된 목록으로 환경 복원 (다른 사람이 실행)
# -r 은 파일을 읽어서 그 안의 패키지를 전부 설치하라는 뜻
pip install -r requirements.txt
```

`requirements.txt` 파일을 열어보면 이런 내용이 들어있습니다:
```
requests==2.31.0
urllib3==2.1.0       # requests가 내부적으로 사용하는 패키지 (자동 설치됨)
certifi==2024.2.2    # requests가 내부적으로 사용하는 패키지 (자동 설치됨)
```

> `requests` 하나만 설치했는데 왜 3개가 있을까?  
> 패키지는 다른 패키지에 **의존**하는 경우가 많습니다.  
> `pip install requests`를 하면 requests가 필요로 하는 패키지들도 자동으로 함께 설치됩니다.
