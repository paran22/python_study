## Part 0-2. 파일 분리와 `import`

### 왜 파일을 나누나요?

코드가 길어지면 하나의 파일에 모든 것을 담기 어렵습니다.  
**역할별로 파일을 나누고**, 필요한 것만 가져다 쓰는 방식을 사용합니다.

예를 들어 나중에 에이전트를 만들 때:
- `main.py` — 실행
- `search.py` — 검색 기능
- `llm.py` — AI 요약 기능

### 파일별 스코프

1차시에서 함수 안과 밖의 변수가 분리되는 것(스코프)을 배웠습니다.  
**파일이 다르면 변수와 함수도 서로 보이지 않습니다.**  
다른 파일의 함수를 쓰려면 `import`로 가져와야 합니다.

### `import` 두 가지 방식

```python
# 방법 1: 특정 함수만 가져오기
from prompts import make_prompt
result = make_prompt("영양사", "다이어트 식단")

# 방법 2: 파일 전체를 가져오고, 파일이름.함수이름()으로 사용
import prompts
result = prompts.translate_prompt("오늘 날씨가 좋다", "한국어", "영어")
```

### 실습 안내

이 파트는 **Cursor에서 진행**합니다.  
`2차시_import실습/` 폴더의 `prompts.py`와 `main.py`를 열고,  
`main.py`의 Step을 하나씩 주석 해제하면서 터미널에서 `python main.py`로 실행해보세요.

## Part 2. 딕셔너리 (Dictionary)

딕셔너리는 **이름표(키)로 값을 찾는 데이터 묶음**입니다.  
중괄호(`{}`)로 만들고, `키: 값` 형태로 저장합니다.

````python
# 딕셔너리 만들기
person = {
    "name": "가은",
    "age": 25,
    "city": "서울"
}

print(person)
````

````python
# 키로 값 꺼내기
print(person["name"])   # "가은"
print(person["age"])    # 25
print(person["city"])   # "서울"
````

````python
# 없는 키를 쓰면? → KeyError
print(person["email"])
````

````python
# 딕셔너리에 어떤 키가 있는지 확인하기
print(person.keys())
````

### 함수 vs 메서드

지금까지 `fruits.append("포도")`나 `person.keys()`처럼 **점(`.`)을 찍어서 호출**하는 것을 봤습니다.  
이것이 바로 **메서드(Method)**입니다.

| 구분 | 형태 | 예시 |
|------|------|------|
| **함수** | `함수이름()` | `print("안녕")`, `len(fruits)` |
| **메서드** | `데이터.메서드이름()` | `fruits.append("포도")`, `person.keys()` |

- 함수는 **혼자서** 호출합니다: `len(fruits)`
- 메서드는 **데이터에 점을 찍어서** 호출합니다: `fruits.append("포도")`
- 메서드는 "이 데이터가 가지고 있는 기능"이라고 생각하면 됩니다
  - `fruits.append("포도")` → "fruits야, 포도를 추가해"
  - `person.keys()` → "person아, 네가 가진 키 목록 보여줘"

앞으로 API를 다룰 때도 이 패턴이 계속 나옵니다:  
`response.json()`, `text.strip()`, `llm.invoke()` 등 전부 같은 방식!

````python
# 함수: 이름만으로 호출
print(len(fruits))      # len()은 함수

# 메서드: 데이터에 점(.)을 찍어서 호출
fruits.append("멜론")   # .append()는 리스트의 메서드
print(fruits)

print(person.keys())    # .keys()는 딕셔너리의 메서드
````

````python
# ✏️ 직접 해보기
# 나를 소개하는 딕셔너리를 만들고 이름을 출력해보세요

me = {
    "name": "",       # 내 이름
    "hobby": "",      # 내 취미
    "language": ""    # 주로 쓰는 언어
}

print(me[""])          # 이름을 꺼내는 키를 넣어보세요
````

### 딕셔너리를 배우는 이유 — JSON

딕셔너리 문법이 익숙해지면 좋은 이유가 있습니다.  
프로그래밍에서 가장 많이 만나는 데이터 형태 중 하나인 **JSON**이 딕셔너리와 거의 같은 모양이기 때문입니다.

**JSON**(JavaScript Object Notation)은 프로그램끼리 데이터를 주고받을 때 쓰는 표준 형식입니다.  
API 응답, 설정 파일, 데이터 저장 등 거의 모든 곳에서 쓰입니다.

````python
# 이것은 파이썬 딕셔너리입니다
person = {
    "name": "가은",
    "age": 25,
    "city": "서울"
}
print(person)
print(type(person))  # <class 'dict'>
````

````python
# 이것은 JSON 문자열입니다 — 생김새가 딕셔너리와 거의 같습니다
import json

json_text = '{"name": "가은", "age": 25, "city": "서울"}'
print(json_text)
print(type(json_text))  # <class 'str'> — 아직은 그냥 문자열!
````

````python
# json.loads()로 JSON 문자열 → 딕셔너리로 변환
data = json.loads(json_text)
print(data)
print(type(data))     # <class 'dict'> — 이제 딕셔너리!
print(data["name"])   # 딕셔너리니까 ["키"]로 꺼낼 수 있음
````

````python
# 반대로 딕셔너리 → JSON 문자열로 변환
person = {"name": "가은", "age": 25, "city": "서울"}
json_output = json.dumps(person, ensure_ascii=False)
print(json_output)
print(type(json_output))  # <class 'str'>
````

정리하면:

| | JSON | 딕셔너리 |
|---|------|----------|
| **정체** | 텍스트(문자열) | 파이썬 데이터 |
| **용도** | 프로그램 간 데이터 전달 | 파이썬 안에서 데이터 다루기 |
| **변환** | `json.loads()` → 딕셔너리 | `json.dumps()` → JSON |

API를 호출하면 JSON으로 응답이 오고, 이걸 딕셔너리로 바꿔서 `["키"]`로 값을 꺼냅니다.  
그래서 **딕셔너리를 잘 다루면 API 데이터도 바로 다룰 수 있습니다.**

## Part 4. 반복문 (`for`)

`for`는 리스트의 항목을 **하나씩 꺼내서 반복 실행**할 때 사용합니다.

````python
# 리스트의 항목을 하나씩 꺼내기
fruits = ["사과", "바나나", "딸기"]

for fruit in fruits:
    print(fruit)
````

````python
# 먼저 리스트의 길이를 확인해봅시다
fruits = ["사과", "바나나", "딸기"]

print(len(fruits))  # 3개
````

````python
# range는 숫자를 순서대로 만들어줍니다
# range(3)은 0, 1, 2를 만들어줍니다 (3은 포함하지 않음!)
for i in range(3):
    print(i)
````

````python
# i에 들어간 숫자를 인덱스로 쓸 수 있습니다
# range에 len(fruits)를 넣으면 리스트의 모든 인덱스가 순서대로 돌아갑니다
fruits = ["사과", "바나나", "딸기"]

for i in range(len(fruits)):
    print(f"fruits[{i}] → {fruits[i]}")
````

````python
# enumerate — 인덱스와 값을 동시에 꺼내기 (더 깔끔한 방법)
fruits = ["사과", "바나나", "딸기"]

for i, fruit in enumerate(fruits):
    print(f"{i}번째: {fruit}")
````

````python
# 리스트 안의 딕셔너리를 for로 순회하기
movies = [
    {"title": "인셉션", "rating": 9.0},
    {"title": "인터스텔라", "rating": 9.5},
    {"title": "다크나이트", "rating": 9.2}
]

# i는 0부터 시작하므로, 사람이 읽기 편한 번호(1부터)를 보여주려면 i+1을 씁니다
for i, movie in enumerate(movies):
    print(f"인덱스: {i} → [{i+1}] {movie['title']} — {movie['rating']}점")
````

## Part 5. 조건문 (`if`)

`if`는 **조건이 참일 때만 실행**되는 코드를 작성할 때 사용합니다.

````python
# 기본 if 구문
score = 85

if score >= 90:
    print("A등급")
elif score >= 80:
    print("B등급")
else:
    print("C등급")
````

````python
# 문자열 안에 특정 단어가 포함되어 있는지 확인
review = "이 영화 정말 재미있다"

if "재미" in review:
    print("긍정적인 리뷰입니다")
else:
    print("판단하기 어렵습니다")
````

````python
# ✏️ 직접 해보기
# 빈칸을 채워서 temperature가 1.0보다 크면 "창의적", 아니면 "보수적"을 출력해보세요

temperature = 0.7

if temperature   :       # 조건을 채워보세요
    print("창의적")
else:
    print("보수적")
````

### 조건 여러 개 조합하기 — `and`, `or`

조건이 하나만 필요할 때도 있지만, **여러 조건을 동시에 확인**해야 할 때가 많습니다.

| 연산자 | 뜻 | 참이 되려면 |
|--------|-----|------------|
| `and` | 그리고 | 양쪽 **모두** 참 |
| `or` | 또는 | 둘 중 **하나만** 참 |

````python
# and — 두 조건이 모두 참일 때만 실행
score = 85
attendance = 90

if score >= 80 and attendance >= 80:
    print("합격")
else:
    print("불합격")
````

````python
# 하나라도 조건을 못 채우면?
score = 85
attendance = 70  # 출석이 80 미만

if score >= 80 and attendance >= 80:
    print("합격")
else:
    print("불합격")  # and는 둘 다 참이어야 하므로 여기가 실행됨
````

````python
# or — 둘 중 하나만 참이어도 실행
has_ticket = False
is_vip = True

if has_ticket or is_vip:
    print("입장 가능")  # 티켓은 없지만 VIP이므로 입장 가능
else:
    print("입장 불가")
````

````python
# and와 or를 함께 쓰기
# 예: 영화 평점이 9.0 이상이고, 장르가 액션이거나 SF인 경우
rating = 9.5
genre = "SF"

if rating >= 9.0 and (genre == "액션" or genre == "SF"):
    print(f"{genre} 장르, 평점 {rating} — 추천!")
else:
    print("조건에 맞지 않습니다")
````

````python
# ✏️ 직접 해보기
# 빈칸을 채워서 나이가 13 이상이고 19 이하일 때 "청소년"을 출력해보세요

age = 16

if age >= 13      age <= 19:  # 어떤 연산자를 넣어야 할까요?
    print("청소년")
else:
    print("청소년이 아닙니다")
````

### `if`로 에러 미리 막기 (예외처리)

코드를 실행하다 보면 예상치 못한 상황이 생깁니다.
- 딕셔너리에 **없는 키**로 접근 → `KeyError`
- 빈 리스트에서 **[0]으로 접근** → `IndexError`
- 값이 **None**인데 그대로 사용 → `TypeError`

이런 에러를 미리 `if`로 확인하고 피하는 것을 **예외처리**라고 합니다.  
특히 외부 API에서 데이터를 받아올 때, 값이 빠져있거나 비어있는 경우가 자주 있어서 이 패턴이 중요합니다.

````python
# 딕셔너리에 키가 있는지 확인하기
# 없는 키로 접근하면 KeyError가 나니까, 먼저 확인하면 안전합니다
person = {"name": "가은", "age": 25}

if "email" in person:
    print(person["email"])
else:
    print("email 키가 없습니다")
````

````python
# 키가 있을 때만 꺼내기 — 실전에서 자주 쓰는 패턴
person = {"name": "가은", "age": 25, "email": "gaeun@example.com"}

if "email" in person:
    print(person["email"])
else:
    print("email 키가 없습니다")
````

````python
# 리스트가 비어있는지 확인하기
# 빈 리스트에서 [0]으로 접근하면 IndexError가 납니다
results = []

if len(results) > 0:
    print(f"첫 번째 결과: {results[0]}")
else:
    print("결과가 없습니다")
````

````python
# 결과가 있을 때는 정상 동작
results = ["파이썬 입문 가이드", "파이썬 자료구조 정리"]

if len(results) > 0:
    print(f"첫 번째 결과: {results[0]}")
else:
    print("결과가 없습니다")
````

````python
# 값이 None인지 확인하기
# API 응답에서 값이 비어있을 때 None이 오는 경우가 많습니다
document = None

if document is not None:
    print(document)
else:
    print("내용이 비어있습니다")
````

````python
# ✏️ 직접 해보기
# data 딕셔너리에 "title" 키가 있으면 출력하고, 없으면 "제목 없음"을 출력해보세요

data = {"url": "https://example.com", "content": "파이썬 입문 가이드"}

if "" in data:          # 어떤 키를 확인해야 할까요?
    print(data[""])
else:
    print("제목 없음")
````

## ✏️ 종합 실습: 나만의 키워드로 리뷰 분석하기

오늘 배운 딕셔너리 파싱 + for + if를 모두 활용해서,  
원하는 키워드가 포함된 리뷰를 긍정/부정으로 분류해봅시다.

````python
# 예시: "연기"가 포함된 리뷰를 긍정/부정으로 나누어 세기
keyword = "연기"
positive_count = 0
negative_count = 0

for item in dataset[:5000]:  # 앞의 5000개만 확인
    if item["document"] is not None and keyword in item["document"]:
        if item["label"] == 1:
            positive_count = positive_count + 1
        else:
            negative_count = negative_count + 1

print(f"'{keyword}' 포함 리뷰 분석 결과:")
print(f"  긍정: {positive_count}개")
print(f"  부정: {negative_count}개")
````
